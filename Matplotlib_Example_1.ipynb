```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

style.use('tableau-colorblind10')
# Solarize_Light2, bmh, fivethirtyeight, seaborn, tableau-colorblind10
```

```
df = pd.read_csv('/content/pricesss.csv', encoding='utf-8', parse_dates=True, index_col=['Date'])
df.index = pd.to_datetime(df.index).strftime('%d-%b-%Y')
df.head()
```

```
df1 = df.iloc[: 15]
df2 = df.iloc[16: ]
```

```
plt.subplots(figsize=(15, 5))
plt.plot(df1, '--', alpha = 0.5, linewidth=2)
plt.plot(df2, linewidth=2)
plt.xlabel('date')
plt.ylabel('price')
plt.xticks(df.index, rotation=90) 
plt.show()
```
